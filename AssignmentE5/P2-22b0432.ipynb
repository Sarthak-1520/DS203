```python
# Calculate the correlation matrix
corr = df_new.corr()

# Create a heatmap
sns.heatmap(corr, fmt=".2f")
plt.show()
```

```python
def plot_heatmap(corr):   
    # Create a mask for the upper triangle
    mask = np.triu(np.ones_like(corr, dtype=bool))

    # Set up the matplotlib figure
    f, ax = plt.subplots(figsize=(11, 9))

    # Generate a custom diverging colormap
    cmap = sns.diverging_palette(230, 20, as_cmap=True)

    # Draw the heatmap with the mask and correct aspect ratio
    sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
                square=True, linewidths=.5, cbar_kws={"shrink": .5})

    plt.show()
    
plot_heatmap(corr)
```

```python
from scipy.stats import shapiro
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
def dis_check(df):
    df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
    sum = 0 
    # Perform Shapiro-Wilk test on each column
    for column in df_normalized.columns:
        stat, p = shapiro(df_normalized[column])
        print(f'Column {column}: Statistics={stat}, p={p}')
        alpha = 0.05
        if p > alpha:
            sum += 1 
            print('Sample looks Gaussian (fail to reject H0)')
        else:
            print('Sample does not look Gaussian (reject H0)')

dis_check(df_new)

        
print(f"Number of columns that are normally distributed : {sum}")
print(f"Number of columns that are not normally distributed : {99-sum}")
```

```python
def IQR_Out_Check(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))].count()
    return outliers.sum()
IQR_Out_Check(df_new)
```

# Total number of outliers = 2775 as per IQR method
# We select IQR method as we know by p-value statistics (p < 0.05) that the data distributed atleast for the most part is not normally distributed.

```python
Q1 = df_new.quantile(0.25)
Q3 = df_new.quantile(0.75)
IQR = Q3 - Q1
# define the upper and lower bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# remove the outliers from the dataframe
df_no_outliers = df_new[~((df_new < lower_bound) | (df_new > upper_bound)).any(axis=1)]
IQR_Out_Check(df_no_outliers)
```

# Thus, we have significatly reduced the Number of outliers from 2775 --> 866 

```python
from scipy.stats import shapiro
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df_no_outliers), columns=df_no_outliers.columns)
sum = 0 
# Perform Shapiro-Wilk test on each column
for column in df_normalized.columns:
    stat, p = shapiro(df_normalized[column])
    print(f'Column {column}: Statistics={stat}, p={p}')
    alpha = 0.05
    if p > alpha:
        sum += 1 
        print('Sample looks Gaussian (fail to reject H0)')
    else:
        print('Sample does not look Gaussian (reject H0)')
        
print(f"Number of columns that are normally distributed : {sum}")
print(f"Number of columns that are not normally distributed : {99-sum}")
```

```python
from scipy.stats import normaltest
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
standard_scaler = StandardScaler()

# Loop through each column in the dataframe
for column in df_no_outliers.columns:
    # Check if the p-value of the column is greater than 0.05
    if normaltest(df_no_outliers[column]).pvalue > 0.05:
        # Normalize the column
        df_no_outliers[column] = (df_no_outliers[column] - df_no_outliers[column].mean()) / df_no_outliers[column].std()
    else:
        # Standardize the column
        df_no_outliers[column] = standard_scaler.fit_transform(df_no_outliers[column].values.reshape(-1, 1))

df_no_outliers.head()
```
