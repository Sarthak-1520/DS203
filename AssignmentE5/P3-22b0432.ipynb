Since it is a image based data we should Standardise the data to exploit the sparsity (significant number of values = 0 ) of the data. 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Standardize the data
scaler = StandardScaler()
df_std = scaler.fit_transform(df)

# Apply PCA for 3 components
pca_3 = PCA(n_components=3)
df_pca = pca_3.fit_transform(df_std)

# Create a DataFrame for the reduced data
df_pca = pd.DataFrame(data=df_pca, columns=['PC1', 'PC2', 'PC3'])

# Concatenate the labels to the DataFrame
df_pca = pd.concat([df_pca, df[['label']]], axis=1)

# Plot the data in 3D space
fig = px.scatter_3d(df_pca, x='PC1', y='PC2', z='PC3', color='label')
fig.update_layout(title='MNIST digits in 3D space', autosize=False, width=800, height=600)
fig.show()

```

```python
# Apply PCA for 3 components
pca_2 = PCA(n_components=2)
df_pca = pca_2.fit_transform(df_std)

# Create a DataFrame for the reduced data
df_pca = pd.DataFrame(data=df_pca, columns=['PC1', 'PC2'])

# Concatenate the labels to the DataFrame
df_pca = pd.concat([df_pca, df[['label']]], axis=1)

# Plot the data in 3D space
fig = px.scatter(df_pca, x='PC1', y='PC2', color='label')
fig.update_layout(title='MNIST digits in 2D space', autosize=False, width=800, height=600)
fig.show()
```

```python
import plotly.graph_objects as go

# Apply PCA with varying number of components
n_components = range(1, 51)  # Change this to the range you want
explained_variance = []

for i in n_components:
    pca = PCA(n_components=i)
    pca.fit(df_std)
    explained_variance.append(np.sum(pca.explained_variance_ratio_))

# Plot the elbow diagram using Plotly
fig = go.Figure(data=go.Scatter(x=list(n_components), y=explained_variance, mode='lines+markers'))
fig.update_layout(title='Elbow Diagram',
                   xaxis_title='Number of Components',
                   yaxis_title='Cumulative Explained Variance')
fig.show()
```

```python
from sklearn.manifold import TSNE
import plotly.graph_objects as go

# Initialize t-SNE
tsne = TSNE(n_components=2, random_state=0)

# Fit and transform the data
tsne_results = tsne.fit_transform(df_std)

# Create a scatter plot of the t-SNE results using Plotly
fig = go.Figure()

for label in df['label'].unique():
    idx = df['label'] == label
    fig.add_trace(go.Scatter(x=tsne_results[idx, 0], y=tsne_results[idx, 1], mode='markers', 
                             name=str(label)))

fig.update_layout(title='t-SNE plot',
                   xaxis_title='Dimension 1',
                   yaxis_title='Dimension 2')
fig.show()
```
